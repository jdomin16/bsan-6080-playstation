```python
#Dropping columns with null values
df_tweets = tweets_data.drop(['near','geo','source','user_rt_id','user_rt','retweet_id','retweet_date','translate','trans_src','trans_dest'], axis=1)
```

```python
top_5_liked_tweets = df_tweets.sort_values(by="likes_count", ascending=False).head() #Getting the top 5 tweets from the data
```

```python
top_5_liked_tweets.head()
```

```python
top_5_liked_tweets.plot(x = "date", y="likes_count", kind="bar", rot=5, fontsize=8) #Plotting the top 5 tweets by date
```
